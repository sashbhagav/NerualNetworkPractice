```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Breast_cancer_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['diagnosis']
X = df.drop(columns=['diagnosis'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
```

```python
n_input = 5 #number of input layers
n_hidden = 10 #number of hidden nodes (1 layer)
n_hidden2 = 8 #number of hidden nodes (1 layer)
n_hidden3 = 6 #number of hidden nodes (1 layer)
n_output = 1
```

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

```python
#assigning weights between layers and giving them initial value 
w1 = tf.Variable(tf.random_uniform([n_input, n_hidden], 0, 1.0))
w2 = tf.Variable(tf.random_uniform([n_hidden, n_hidden2], 0, 1.0))
w3 = tf.Variable(tf.random_uniform([n_hidden2, n_hidden3], 0, 1.0))
w4 = tf.Variable(tf.random_uniform([n_hidden3, n_output], 0, 1.0))
```

```python
#bias term for each layer of NN
b1 = tf.Variable(tf.zeros([n_hidden]), name='Bias1')
b2 = tf.Variable(tf.zeros([n_hidden2]), name='Bias2')
b3 = tf.Variable(tf.zeros([n_hidden3]), name='Bias3')
b4 = tf.Variable(tf.zeros([n_output]), name='Bias4')
```

```python
l2 = tf.nn.softmax(tf.matmul(X, w1) + b1)
l3 = tf.nn.softmax(tf.matmul(l2, w2) + b2)
l4 = tf.nn.softmax(tf.matmul(l3, w3) + b3)
hy = tf.nn.softmax(tf.matmul(l4, w4) + b4)
```

```python
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(hy), reduction_indices=[1]))
```

```python
learning_rate=.01
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
init = tf.initialize_all_variables()
with tf.Session() as session:
    session.run(init)
    
    
    epochs = 10000
    for i in range(epochs):
        session.run(optimizer, feed_dict={X: X_train, Y: y_train})
        
        if i % 1000 == 0:
            print(session.run(cost, feed_dict={X: X_train, Y: y_train}))
    answer = tf.equal(tf.floor(hy + .5), Y)
    accuracy = tf.reduce_mean(tf.cast(answer, "float"))
    
    print(session.run([hy], feed_dict={X: X_train, Y: y_train}))
    print("accuracy: ", accuracy.eval({X: X_train, Y: y_train}))
    
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Breast_cancer_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['diagnosis']
X = df.drop(columns=['diagnosis'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
n_input = df.shape[1] #number of input layers
n_hidden = 10 #number of hidden nodes (1 layer)
n_output = 1
```

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

```python
#assigning weights between layers and giving them initial value 
w1 = tf.Variable(tf.random_uniform([n_input, n_hidden], -1.0, 1.0))
w2 = tf.Variable(tf.random_uniform([n_hidden, n_output], -1.0, 1.0))
```

```python
#bias term for each layer of NN
b1 = tf.Variable(tf.zeros([n_hidden]), name='Bias1')
b2 = tf.Variable(tf.zeros([n_output]), name='Bias2')
```

```python
l2 = tf.sigmoid(tf.matmul(X, w1) + b1)
hy = tf.sigmoid(tf.matmul(l2, w2) + b2)
```

```python
cost = tf.reduce_mean(-Y*tf.log(hy) - (1-Y)*tf.log(1-hy))
```

```python
learning_rate=.01
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```
